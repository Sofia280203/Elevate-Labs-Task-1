```python
import pandas as pd
df = pd.read_csv('marketing_campaign.csv',sep='\t')
df.head()
```

```python
#checking for and handling missing values
missing_vals = df.isnull().sum()
print('Missing Values: ', missing_vals)
```

```python
df = df.dropna()
df.head()
```

```python
df.shape[0]
```

```python
#Dropping Duplicates
df_clean = df.drop_duplicates()
df_clean
```

```python
#standardising text values and renaming columns for consistency
df_clean['Education'] = df_clean['Education'].str.strip().str.capitalize()
df_clean['Marital_Status'] = df_clean['Marital_Status'].str.strip().str.capitalize()
df_clean['Dt_Customer'] = pd.to_datetime(df_clean['Dt_Customer'],format='%d-%m-%Y',errors='coerce')
df_clean.columns = df_clean.columns.str.lower().str.strip().str.replace(' ', '_')
df_clean
```

```python
#checking and fixing dtypes
df_clean.dtypes
```

```python
df_clean['income'] = df_clean['income'].astype(int)
df_clean['education'] = df_clean['education'].astype('category')
df_clean['marital_status'] = df_clean['marital_status'].astype('category')
df_clean.dtypes
```

```python
df_clean.to_csv('marketing_campaign_cleaned.csv', index = False)
```
